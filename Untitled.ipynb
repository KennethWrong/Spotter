```python
from textblob import TextBlob
```

```python
def get_sentiment(text):
    blob = TextBlob(text)
    sentiment = blob.sentiment.polarity
    if sentiment > 0:
        result = "Positive"
    elif sentiment < 0:
        result = "Negative"
    else:
        result = "Neutral"
    
    return result
```

```python
get_sentiment("I love coding")
```

```python
df['Sentiment'] = df['Text'].apply(get_sentiment)
```

```python
### Text Cleaning
df['Clean_Text'] = df['Text'].apply(nfx.remove_stopwords)
```

```python
df['Clean_Text'] = df['Clean_Text'].apply(nfx.remove_userhandles)
df['Clean_Text'] = df['Clean_Text'].apply(nfx.remove_punctuations)
```

```python
df[['Text', 'Clean_Text']]
```

```python
### Keyword Extraction
# Extract most commonest words per class of emotion
from collections import Counter
```

```python
def extract_keywords(text, num=50):
    tokens = [token for token in text.split()]
    most_common_tokens = Counter(tokens).most_common(num)
    return dict(most_common_tokens)
```

```python
emotion_list = df["Emotion"].unique().tolist()
print(emotion_list)
```

```python
joy_list = df[df['Emotion'] == 'joy']['Clean_Text'].tolist()

joy_docx = ' '.join(joy_list)
```

```python
#Extract keywords
keyword_joy = extract_keywords(joy_docx)
```

```python
#Plot function
def plot_most_common_words(mydict):
    df_01 = pd.DataFrame(mydict.items(), columns=['token', 'count'])
    plt.figure(figsize=(20, 10))
    sns.barplot(x='token', y='count', data=df_01)
    plt.xticks(rotation=45)
    plt.show()
```

```python
plot_most_common_words(keyword_joy)
```

```python
## Machine Learning

# Load ML Packages
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB

# Vectorizer
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# Metrics
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
from sklearn.model_selection import train_test_split

# Build Features from our text
Xfeatures = df['Clean_Text']
ylabels = df['Emotion']
```

```python
# Vectorizer
cv = CountVectorizer()
X = cv.fit_transform(Xfeatures)
```

```python
X.toarray()
```

```python
#Split Dataset
X_train, X_test, y_train, y_test = train_test_split(X, ylabels, test_size=0.3, random_state=42)
```

```python
# Build model
nv_model = MultinomialNB()
nv_model.fit(X_train, y_train)
```

```python
#Accuracy
nv_model.score(X_test, y_test)
```

```python
y_pred_for_nv = nv_model.predict(X_test)
```

```python
y_pred_for_nv
```

```python
sample_test = ["I just failed my exam"]
vect = cv.transform(sample_test).toarray()
```

```python
prediction = nv_model.predict(vect)

pred_prob=nv_model.predict_proba(vect)
nv_model.classes_

print({x:y for (x,y) in zip(nv_model.classes_, pred_prob[0])})
```

```python
def predict_emotion(text, model):
    vect = cv.transform(text).toarray()
    pred = model.predict(vect)
    pred_proba = model.predict_proba(vect)
    pred_percentage_for_all = dict(zip(model.classes_, pred_proba))
    print(pred[0])
    return pred_percentage_for_all
```

```python
sample_test = ["I just got engaged, I didnt even realize it"]
predict_emotion(sample_test, nv_model)
```

```python
print(classification_report(y_test, y_pred_for_nv))
```

```python
confusion_matrix(y_test, y_pred_for_nv)
```

```python
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
```

```python
predict_emotion(sample_test, lr_model)
lr_model.score(X_test, y_test)
```
